# 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos


```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
from selenium.webdriver.common.by import By
```

```python
YouTubeVideos = []

table = driver.find_elements(By.XPATH,"//*[@id='mw-content-text']/div[1]/table[2]/tbody")

for row in table:
    #print(row.find_elements(By.TAG_NAME,"td"))
    cols = row.find_elements(By.TAG_NAME,"tr")
    for td in cols:
        cells = td.find_elements(By.TAG_NAME,"td")        
        data = {
            "Rank":cells[0].text,
            "Name":cells[1].text,
            "Artist":cells[2].text,
            "Upload Date":cells[3].text,
            "Views":cells[4].text
            
        }
        YouTubeVideos.append(data)

for name in YouTubeVideos:
    print(name['Rank'],name['Name'],name['Artist'],name['Upload Date'],name['Views'])
    
df=pd.DataFrame(YouTubeVideos)     
```

# 5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/

```python
driver.get("https://github.com/")
```

```python
WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a"))).click()
```

```python
body = driver.find_element(By.XPATH,"/html/body/div[4]/main/div[3]")
repositories=[]
products = driver.find_elements(By.CLASS_NAME,'Box-row')
for product in products:
    #print(product.find_element(By.XPATH,'//span[@itemprop="programmingLanguage"]').text)    
    try:
        RepositoryDescription = product.find_element(By.CLASS_NAME, 'col-9.color-fg-muted').text
    except Exception as ex:
        RepositoryDescription = '-'
    Contributors = product.find_elements(By.CLASS_NAME, 'Link--muted.d-inline-block')   
    try:
        ContributorsCount = Contributors[1].text
    except Exception as ex:
        ContributorsCount = '-'    
    data =  { 
        "Repository title":product.find_element(By.CLASS_NAME,"text-normal").text.replace('/',''),
        "Repository description":RepositoryDescription,
        "Contributors count":ContributorsCount,
        "Language used": product.find_element(By.XPATH,'//span[@itemprop="programmingLanguage"]').text

        
    }
    repositories.append(data)
    
for name in repositories:
    print(name['Repository title'],name['Repository description'],name['Contributors count'],name['Language used'])
```

```python
body = driver.find_element(By.XPATH,"//*[@id='main']/div/div[3]")
series=[]
products = driver.find_elements(By.CLASS_NAME,'lister-item-content')
for product in products:
    Votes = product.find_elements(By.CLASS_NAME,"text-muted.text-small")
    try:
        Vote = Votes[2].text
    except Exception as ex:
        vote = '-'
    data =  { 
        "Name":product.find_element(By.TAG_NAME,"a").text,
        "Year":product.find_element(By.CLASS_NAME,"lister-item-year.text-muted.unbold").text,
        "Genre":product.find_element(By.CLASS_NAME,"genre").text,
        "RunTime":product.find_element(By.CLASS_NAME,"runtime").text,
        "Ratings":product.find_element(By.CLASS_NAME,"ipl-rating-star__rating").text,
        "Votes":Vote
    }
    series.append(data)
    
for name in series:
    print(name['Name'],name['Year'],name['Genre'],name['RunTime'],name['Ratings'],name['Votes'])
```

# 10. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/

```python
driver.get("https://archive.ics.uci.edu/")
```

```python
WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "/html/body/table[1]/tbody/tr/td[2]/span[2]/a"))).click()
```

```python
dataset=[]
table = driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody")
for row in table:
    cols = row.find_elements(By.TAG_NAME,"tr")
    for td in cols:
        cells = td.find_elements(By.TAG_NAME,"td") 
        try :
            DataTypeValue = cells[3].text
        except Exception as ex:
            DataTypeValue = '-'
        try :
            TaskValue = cells[2].text
        except Exception as ex:
            TaskValue = '-'
        try :
            AttributeValue = cells[3].text
        except Exception as ex:
            AttributeValue = '-' 
        try :
            InstancesValue = cells[4].text
        except Exception as ex:
            InstancesValue = '-'
        try :
            AttributeNoValue = cells[5].text
        except Exception as ex:
            AttributeNoValue = '-'
        try :
            YearValue = cells[6].text
        except Exception as ex:
            YearNoValue = '-'    
        data = {
            "Name":cells[0].text,
            "DataType":DataTypeValue,
            "Task":TaskValue,
            "Attribute":AttributeValue,
            "Instances":InstancesValue,
            "No of Attribute":AttributeNoValue,
            "Year":YearValue
        }
        dataset.append(data)
        
for name in dataset:
    name['Name'],name['DataType'],name['Task'],name['Instances'],name['No of Attribute'],name['Year']
    
df=pd.DataFrame(dataset)
df
```
